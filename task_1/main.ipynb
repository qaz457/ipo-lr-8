программа берет оценки с файла students.xlsx по физике,информатике и математики и вычесляет среднюю оценку каждого предмета

```python
from openpyxl import load_workbook

wb = load_workbook("data/students.xlsx")
sheet_marks = wb["Оценки"]
score = 0
mark_math = 0
mark_fit = 0
mark_info = 0

title = []
average_mark = []
title_mark = {}
title_mark['Предмет'] = 'Средняя оценка'
for row in sheet_marks['B1':'D1']:
    for cell in row:
        title.append(cell.value)
for row in sheet_marks['B2':'B108']:
    for cell in row:
        mark_math += cell.value
        score += 1
        average_mark_math = round(mark_math/score,2)
average_mark.append(average_mark_math)
for row in sheet_marks['C2':'C108']:
    for cell in row:
        mark_fit += cell.value
        average_mark_fit = round(mark_fit/score,2)
average_mark.append(average_mark_fit)
for row in sheet_marks['D2':'D108']:
    for cell in row:
        mark_info += cell.value
        average_mark_info = round(mark_info/score,2)
average_mark.append(average_mark_info)
for i in range(len(title)):
    title_mark[title[i]] = average_mark[i]
print(f"Средняя оценка по математике:{average_mark_math}")
print(f"Средняя оценка по физике:{average_mark_fit}")
print(f"Средняя оценка по информатики:{average_mark_info}")
print(title_mark)
```

В этой части программы мы постороили столбчатую диограмму по данным из первой части,добавили подписи и сетку

```python
import matplotlib.pyplot as mp
import numpy as np

mp.bar(title,average_mark,label='Средняя оценка')
mp.grid(True)
mp.ylim(0,10)
mp.yticks(range(0,11,1))
mp.xlabel('Предмет')
mp.ylabel('Оценка')
mp.title("Столбачатая диограмма")
mp.legend()
mp.show()

```

Записал данные с графика в файл students_result.xlsx

```python
from openpyxl import Workbook

wb = Workbook()

page = wb.active
page.title = 'Средняя оценка'

for i , (title,mark) in enumerate(title_mark.items(),start = 1):
    page[f'A{i}'] = title
    page[f'B{i}'] = mark

wb.save('data/students_result.xlsx')
```
